## 3. Run Experiment to Generate Results

Run an experiment, the agent will generate outputs for each input and the LLM judge will evaluate each one.

```python
from agent import recipe_bot as agent
from langchain.messages import HumanMessage
from langsmith import Client, evaluate

client = Client()


def run_agent(inputs: dict) -> dict:
    response = agent.invoke(
        {"messages": [HumanMessage(content=inputs["query"])]}
    )
    return {"output": response["messages"][-1].content}

results_train = evaluate(
    run_agent,
    data=client.list_examples(dataset_name=DATASET_NAME),
    experiment_prefix="align"
)
```
